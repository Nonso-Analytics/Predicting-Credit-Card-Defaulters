## Feature Engineering

```python
# Feature Engineering
cc_df['avg_bill_amt'] = cc_df[[f'bill_amt{i}' for i in range(1,7)]].mean(axis=1)
cc_df['avg_pay_amt'] = cc_df[[f'pay_amt{i}' for i in range(1,7)]].mean(axis=1)

cc_df['max_pay_delay'] = cc_df[[f'pay_{i}' for i in [0,2,3,4,5,6]]].max(axis=1)
cc_df['num_late_payments'] = cc_df[[f'pay_{i}' for i in [0,2,3,4,5,6]]].apply(lambda row: (row > 0).sum(), axis=1)
cc_df['avg_pay_delay'] = cc_df[[f'pay_{i}' for i in [0,2,3,4,5,6]]].mean(axis=1)
cc_df['was_ever_late'] = (cc_df[[f'pay_{i}' for i in [0,2,3,4,5,6]]] > 0).any(axis=1).astype(int)
```

```python
# List of individual bill and payment columns to drop
bill_cols_to_drop = [f'bill_amt{i}' for i in range(1, 7)]
pay_cols_to_drop = [f'pay_amt{i}' for i in range(1, 7)]
pay2_cols_to_drop = [f'pay_{i}' for i in [0,2,3,4,5,6]]

# Combine the lists
cols_to_drop = bill_cols_to_drop + pay_cols_to_drop + pay2_cols_to_drop

# Drop the individual bill and payment columns
cc_df.drop(columns=cols_to_drop, axis=1, inplace=True)
```

```python
# Correlation matrix
corr = cc_df.corr(numeric_only=True)
mask = np.triu(np.ones_like(corr, dtype=bool))
plt.figure(figsize=(14, 10))

sns.heatmap(
    corr,
    mask=mask,
    cmap='coolwarm',
    annot=True,
    fmt=".2f",
    linewidths=0.5,
    cbar_kws={"shrink": 0.8}
)

plt.title('Correlation Matrix', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Identify highly correlated features (threshold = 0.95)
threshold = 0.95
correlated_features = set()
for i in range(len(corr.columns)):
    for j in range(i):
        if abs(corr.iloc[i, j]) > threshold:
            colname = corr.columns[i]
            correlated_features.add(colname)

# Remove the highly correlated features
cc_df_reduced = cc_df.drop(correlated_features, axis=1)

# Display the DataFrame after removing highly correlated features
print("Removed Features:", correlated_features)
print("Remaining columns in DataFrame:", cc_df_reduced.columns)
```

## Model Building

```python
cc_df_reduced = cc_df_reduced.drop('id', axis=1)
X = cc_df_reduced.drop('default payment next month', axis = 1)
y = cc_df_reduced['default payment next month']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
from sklearn.decomposition import PCA

# Initialize PCA - Let's start by keeping 95% of the variance
pca = PCA(n_components=0.95, random_state=42)

# Fit PCA on the training data and transform both training and testing data
X_train_pca = pca.fit_transform(X_train_resampled)
X_test_pca = pca.transform(X_test)

# Let's see how many components PCA chose to keep
print(f"PCA selected {pca.n_components_} components.")
```

### **Baseline model with default parameters**

```python
lr_model = LogisticRegression(random_state=42)
rf_model = RandomForestClassifier(random_state=42)

models = [lr_model, rf_model]
```

```python
# function for comparing different models
def compare_models(models, X_train_pca, X_test_pca, y_train_resampled, y_test):
    for model in models:
        model.fit(X_train_pca, y_train_resampled)
        y_pred = model.predict(X_test_pca)
        train_f1 = f1_score(y_train_resampled, model.predict(X_train_pca))
        test_f1 = f1_score(y_test, y_pred)
        test_cm = confusion_matrix(y_test, y_pred)
        print(f"Train_F1 {model.__class__.__name__}: {train_f1}")
        print(f"Test_F1 {model.__class__.__name__}: {test_f1}")
        print(f"Confusion Matrix {model.__class__.__name__}:\n {test_cm}")
        print("================================================================")
```

```python
compare_models(models, X_train_pca, X_test_pca, y_train_resampled, y_test)
```

clear signs of overfitting, lets try different parameters and a boosting algorithm

```python
rf_model = RandomForestClassifier(max_depth=5, min_samples_split=20, min_samples_leaf=10,
                                  class_weight= 'balanced', n_estimators=100, random_state=42)
lr_model = LogisticRegression(C=0.1, class_weight='balanced', random_state=42)
xgb_model = XGBClassifier(max_depth=3, learning_rate=0.1, n_estimators=100, random_state=42)
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
skf = StratifiedKFold(n_splits=5)

# function for comparing different models
models = [lr_model, rf_model, xgb_model]
def compare_models(models, X_train_pca, X_test_pca, y_train_resampled, y_test):
    for model in models:
        model.fit(X_train_pca, y_train_resampled)
        y_pred = model.predict(X_test_pca)
        train_f1 = f1_score(y_train_resampled, model.predict(X_train_pca))
        test_f1 = f1_score(y_test, y_pred)
        scores = cross_val_score(model, X_train_pca, y_train_resampled, cv=skf, scoring='f1_weighted')
        cv_score = scores.mean()
        class_report = classification_report(y_test, y_pred)
        test_cm = confusion_matrix(y_test, y_pred)
        print(f"Train_F1 {model.__class__.__name__}: {train_f1}")
        print(f"Test_F1 {model.__class__.__name__}: {test_f1}")
        print(f"CV_Score {model.__class__.__name__}: {cv_score.mean()}")
        print(f"Classification Report {model.__class__.__name__}\n: {class_report}")
        print(f"Confusion Matrix {model.__class__.__name__}:\n {test_cm}")
        print("================================================================")
```

```python
compare_models(models, X_train_pca, X_test_pca, y_train_resampled, y_test)
```

```python
# Get probabilities for the positive class
y_pred_proba_rf = rf_model.predict_proba(X_test_pca)[:, 1]
y_pred_proba_lr = lr_model.predict_proba(X_test_pca)[:, 1]
y_pred_proba_xgb = xgb_model.predict_proba(X_test_pca)[:, 1]
```

```python
model_preds = {
    'Random Forest': y_pred_proba_rf,
    'Logistic Regression': y_pred_proba_lr,
    'XGBoost': y_pred_proba_xgb,
}
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# ROC Curve
for name, y_proba in model_preds.items():
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    auc_score = roc_auc_score(y_test, y_proba)
    axes[0].plot(fpr, tpr, label=f'{name} (AUC = {auc_score:.3f})')

axes[0].plot([0, 1], [0, 1], 'k--', label='Random')
axes[0].set_xlabel('False Positive Rate')
axes[0].set_ylabel('True Positive Rate')
axes[0].set_title('ROC Curve')
axes[0].legend()
axes[0].grid(True)

# PR Curve
for name, y_proba in model_preds.items():
    precision, recall, _ = precision_recall_curve(y_test, y_proba)
    pr_auc = average_precision_score(y_test, y_proba)
    axes[1].plot(recall, precision, label=f'{name} (PR AUC = {pr_auc:.3f})')

axes[1].set_xlabel('Recall')
axes[1].set_ylabel('Precision')
axes[1].set_title('Precision-Recall Curve')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

### Feature Importance

```python
import pandas as pd

# Get feature importances
importances = xgb_model.feature_importances_

# If using PCA-transformed features
feature_names = [f'PC{i+1}' for i in range(len(importances))]

# Create a DataFrame for easy viewing
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

print(importance_df)
```

```python
top_components = importance_df.head(5)['Feature'].apply(lambda x: int(x.replace('PC', '')) - 1)
component_labels = [f'PC{i+1}' for i in top_components]

# Build a mapping of top 5 features per PC (by absolute weight)
top_feature_dict = {}
for idx, pc_index in enumerate(top_components):
    pc_weights = pca.components_[pc_index]
    abs_weights = np.abs(pc_weights)
    top_indices = abs_weights.argsort()[-5:][::-1]
    top_feature_dict[component_labels[idx]] = {
        original_feature: pca.components_[pc_index][i]
        for i, original_feature in zip(top_indices, X.columns[top_indices])
    }


heatmap_df = pd.DataFrame(top_feature_dict).fillna(0)

#Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_df, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Top 5 Original Features for Top 5 XGBoost-Important PCA Components")
plt.xlabel("Principal Components")
plt.ylabel("Original Features")
plt.tight_layout()
plt.show()
```
